```python
N = 50000

red = .28
blue = .29
green = .3
params = [red, blue, green]

red_data = mc.rbernoulli(params[0], N).astype(float)
blue_data = mc.rbernoulli(params[1], N).astype(float)
green_data = mc.rbernoulli(params[2], N).astype(float)
```

```python
sims = 20000

rd = [1.0]; bd = [1.0]; gd = [1.0]
# -- greedy
counter = 1
for i in range(sims):
    eps_greed()
    pass
rd_greed = rd; bd_greed = bd; gd_greed = gd

```

```python
rd = [1.0]; bd = [1.0]; gd = [1.0]
# -- first
counter = 1
for i in range(sims):
    eps_first()
    counter += 1
    pass
rd_first = rd; bd_first = bd; gd_first = gd
```

```python
rd = [1.0]; bd = [1.0]; gd = [1.0]
vec_avg = np.zeros(sims); vec_bound = np.zeros(sims)
# -- ucb1
counter = 1
for i in range(sims):
    x_avg, x_bound = ucb1()
    vec_avg[i] = max(x_avg)
    vec_bound[i] = max(x_bound)
    counter += 1
    pass
rd_ucb1 = rd; bd_ucb1 = bd; gd_ucb1 = gd
```

```python
def cum_avg(x):
    out = [x[0]]
    counter = 1
    for i in range(1, len(x)):
        out.append(((float(out[i-1]) * counter) + x[i]) / (counter + 1))
        counter += 1
        pass
    return out
```

```python
rd_greed_avg = cum_avg(rd_greed)
bd_greed_avg = cum_avg(bd_greed)
gd_greed_avg = cum_avg(gd_greed)

rd_first_avg = cum_avg(rd_first)
bd_first_avg = cum_avg(bd_first)
gd_first_avg = cum_avg(gd_first)

rd_ucb1_avg = cum_avg(rd_ucb1)
bd_ucb1_avg = cum_avg(bd_ucb1)
gd_ucb1_avg = cum_avg(gd_ucb1)
```
